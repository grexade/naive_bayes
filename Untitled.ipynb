```python
x = dataR.drop(columns='Outcome',axis=1)
y = dataR['Outcome']
```

Standardize the Data

```python
scaler = StandardScaler()
scaler.fit(x)
standardized_data = scaler.transform(x)
print(standardized_data)
```

To descale the data: Just for understanding purpose. Not required for this problem.

```python
descale = scaler.inverse_transform(standardized_data)
print(descale)
```

```python
x = standardized_data
y = dataR['Outcome']
print(x)
print(y)
```

Splitting the Dataset into training and testing

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state=2)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
y_train.unique()
```

```python
y_test.unique()
```

Initialize the Naive Bayes classifier

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

Model evaluation using Accuracy score

```python
from sklearn.metrics import accuracy_score
```

```python
x_train_prediction = gnb.predict(x_train)
training_data_accuracy = accuracy_score(y_train , x_train_prediction)
training_data_accuracy
```

```python
print('Accuracy on training data :',training_data_accuracy)
```

Evaluating accuracy of the classifier

```python
x_test_prediction = gnb.predict(x_test)
test_data_accuracy = accuracy_score(y_test , x_test_prediction)
print('Accuracy on testing data :', test_data_accuracy)
```

```python
finalt = np.array(confusion_matrix(y_test, x_test_prediction))
print(finalt)
```

```python
sns.heatmap(finalt, annot= True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
